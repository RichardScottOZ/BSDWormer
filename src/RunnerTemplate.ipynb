This technique works with both gravity and magnetic data.

That being said, its native mathematical environment is gravity (with monopole sources) rather than magnetics (with dipole sources).

In order to use this technique with magnetic data, please use a *pseudogravity transfomed* version of your data as input. The commercial software package Oasis/Montaj has good algorithms for this transformation from Total Magnetic Intensity (TMI). (I believe that their algorithms are even stable for data near the magnetic equator, which is known to cause trouble.)

Should you happen to already have Reduced to the Pole (RTP) magnetic data instead of TMI, recall that the definition of the pseudogravity transform is the vertical integral of RTP data. Oasis/Montaj also has a vertical integral operator, so simply using that on RTP data should suffice to give a pseudogravity dataset.

You need to point to two GIS rasters prepared externally to this code:

1. A base raster (gravity or pseudogravity) from which this code deduces things like pixel size, coordinate reference systems, etc. 
1. A padded version of the information in the base raster, suitable for Fourier Domain operations. (See the comment below for some hints on what tapering function I use.) I like to taper to something like 10-20% of the raster size, so that the sharp edges at the boundary have a more-or-less tamed Fourier Domain representation.

```python
# EXECUTE THIS CELL SECOND AFTER MODIFYING THE VALUES TO SUIT YOUR WORMING JOB.

# This next executable line points to the (unpadded) base GIS raster. This raster
# sets the size, clip outline, and coordinate system of the worming job.
# Since gradients lie at the heart of the worming technique, using a GIS
# to convert to some relevant UTM zone (or other coordinate system that expresses 
# positions in meters) is highly recommended. If you use lat/lon coordinates
# in this raster, gradients are not evaluated correctly (d/d-angle instead of 
# d/d-distance, and the longitudinal distance varies with latitude). 
# USE UTM COORDINATES!
#
# Although in principle, any GIS raster readable by the gdal library should work,
# in practice I've encountered bugs with the stuff exported from Oasis/Montaj. 
# ERMapper .ers raster files definitely work though, so that is recommended.
#
# Modify the string to point to the relevant file for your worming job.
# 
# N.B. The files I'm pointing to here in the template violates the UTM recommendation
# It's in lat/lon coordinates, so gradients are actually off.
# These files just serve as an example that it's perfectly possible to worm something
# in lat/lon -- the code does not care. However you really, really should follow the
# above advice and use UTM.

filename = os.path.abspath('../test_data/SuratBasin/surat_basin.ers')

# This points to the padded GIS raster. Padding is the preparation for Fourier Domain
# operations. Even if you are worming a gravity raster, you still need to pad it since
# upward continuation is expressed in the Fourier Domain.
#
# This image should have zero mean (before the pad) and taper off 
# slowly to zero around the edges.
# I normally use a cosine bell ("Hann" or "Hanning") taper to 
# zero values. 
# I personally use Oasis/Montaj's magmap tools to do the job.
# The coordinate reference system of this file MUST match that of the base file above.
#
# Modify the string to point to the relevant file for your worming job.

padded_filename = os.path.abspath('../test_data/SuratBasin/PaddedSuratBasin.ers')

# This parameter is the "no data" value in the GIS rasters. 
# Adjust to suit whatever your rasters use.

no_data_value = -99999


# The worm results are stored in a PostGIS database under multiple tables that refer to each other.
# This sets a basename for those tables so that they can find each other and 
# you can find the relevant data once you've hooked the database up to ArcGIS or QGIS to 
# look at the worm results.
#
# Modify the string to be something unique and recognizable.

db_basename = 'SuratBasinBGA'

# This is the EPSG code for the coordinate reference system used by the GIS rasters defined above.

srid = 4326

# This call sets up the connection between the worming code and the database.
#
# The parameter 'srid' is the EPSG code that matches the GIS grids above. 
# (I might look at automating that in the future, since that info is buried
# in the GIS files.) For now, you'll need to specify it correctly.
# 
# The parameter 'db' is a black magic config string that tells the code to use postgres,
# with an SQL user name of 'wormuser', a password of 'wormuserpass', 
# running on the locally hosted postgres server, on port 5342, using the
# database named 'wormuser'. It is configured to use the setup matching the
# postgres server running in the vagrant virtual machine. 
# Don't modify this unless you know what you are doing.
#
# Finally, the parameter basename uses db_basename defined above for setting up the tables within
# the database defined by the 'db' parameter.

pgw = PGW(srid=srid,db='postgresql://wormuser:wormuserpass@localhost:5432/wormuser',basename=db_basename)

# This defines the number of (vertical) worm levels you wish to calculate. 
# Levels closer to the surface take longer to compute than levels further away because
# there are more segments. 

num_levels = 3

# This defines what level to start worming at.
# Starting at 0 can cause problems, because that is too close to the sources.
# The 'if dz == 0:' clause in the next cell tries to overcome such problems
# by actually worming at 1% (sort of a numerical "epsilon") of the incremental Z distance 
# away from the sources.
# Starting at 1 is recommended, but it's up to you.

bottom_level = 1

# This defines the incremental Z distance between levels.
# The units should be the same as the horizontal units found in the GIS raster. 
# If you followed the advice above to use UTM coordinates, this should be in meters.
# Even though you can consider the worms to be draped underground, this value must
# be POSITIVE, since it represents the upward continuation part of the worming calculation.

delta_z = 0.00833299999999999

```

```python
# EXECUTE THIS CELL AFTER SETTING THE PARAMETERS ABOVE
# 
# This is where the actual worming code is executed.

job = w.Wormer()

job.importGdalRaster(filename)
job.importExternallyPaddedRaster(padded_filename)
# The argument to range is "num_levels + 1" due to the semantics of Python's range function.
for dz in range(bottom_level,num_levels+1):
    if dz == 0:
        dz = 0.01
    dzm = dz * delta_z
    job.wormLevelAsPoints(dz=dzm)
    # Now that we've upward continued, we need to cancel out the dz = 0.01 cruft above
    if dz == 0.01:
        dz = 0
    job.buildWormSegs(dz=dz,clipped=True,nodata_in_worm_image=no_data_value,from_image=False)
    # This call name is slightly misleading. You need to call it for storing in 
    # PostGIS also, not just VTK.
    #
    job.buildLevelForVTK(dz=dz,delta_z_in_units=dzm)
    pgw.addWormsAtHeightToDB(job,dz,srid=srid)
# This thing does some SRID postprocessing on the bloody PostGIS database. 
# Why? Because I couldn't figure out any other way to make it work with geoalchemy2...
pgw.cleanUpDatabase()

# When this cell finishes execution, the results should be stored in the PostGIS database server.
# You might need to tune the parameters above to get what you want.
#
```
